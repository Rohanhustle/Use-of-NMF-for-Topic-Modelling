```python
import pandas as pd
```

```python
npr=pd.read_csv('../05-Topic-Modeling/npr.csv')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')
```

```python
dtm=tfidf.fit_transform(npr['Article'])
```

```python
from sklearn.decomposition import NMF
```

```python
nmf_model=NMF(n_components=7,random_state=42)
```

```python
nmf_model.fit(dtm)
```

```python
tfidf.get_feature_names()[2300]
```

```python
for index,topic in enumerate(nmf_model.components_):
    print(f"The top 15 words for topic #{index}")
    print([tfidf.get_feature_names()[i] for i in topic.argsort()[-15:]])
    print('\n')
```

```python
topic_results=nmf_model.transform(dtm)
```

```python
topic_results.argmax(axis=1)
```

```python
npr['Topic']=topic_results.argmax(axis=1)
```

```python
mytopic_dict={0:'Health',1:'Election',2:'Legislation',3:'International',4:'Election',5:'Music',6:'Education'}
npr['Topic Label']=npr['Topic'].map(mytopic_dict)
```
